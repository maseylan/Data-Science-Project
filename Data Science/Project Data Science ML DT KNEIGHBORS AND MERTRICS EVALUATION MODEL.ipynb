# MODEL MACHINE LEARNING


```python
#Function untuk df menjadi array untuk dilakukan scalling
def select_columns(df: pd.DataFrame, features: list, target: list = 'rating') : #set default parameter

  x = df[features]
  y = (df[target] >= 3.5).astype(int) #set dtype int

  return x,y
```

```python
#Import train_test
from sklearn.model_selection import train_test_split
```

```python
pip --version
```

```python
features = ['budget','popularity','runtime','vote_average']
x,y = select_columns(data_profit_rating,features = features)
#Menggunakan Function yang sudah dibuat
```

```python
x,y
```

```python
y.value_counts().plot(kind = 'pie',autopct='%1.1f%%')
centre_circle = plt.Circle((0,0),0.65,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Perbandingan Rating >=3.5  ')
plt.figure(figsize=(10, 8))
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3 , random_state = 20)
#random_state untuk memastikan bahwa proses pengacakan antara satu komputer dengan yg lain menghasilkan output yang sama
```

```python
#Import Scaler
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax = MinMaxScaler()
```

```python
minmax.fit(x_train)
```

```python
x_train_scaled = minmax.transform(x_train)
x_test_scaled = minmax.transform(x_test)
```

```python
#function array transform menjadi dataframe
def scaling (df,scaler) :
  scaled_array = scaler.transform(df)
  scaled_df = pd.DataFrame(scaled_array,columns = minmax.get_feature_names_out())

  return scaled_df
```

```python
x_train_scaled = scaling(x_train,minmax)
x_test_scaled = scaling(x_test,minmax)
```

```python
x_train_scaled.head()
```

```python
x_test_scaled
```

```python
#Import Model Machine Learning
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
#import DTree classifier
#import KNeigborsClassifier
```

```python
#K Nearest Neighbor , K sebagai hyper parameter
knn = KNeighborsClassifier(n_neighbors = 3)
dt = DecisionTreeClassifier()
```

```python
knn.fit(x_train_scaled, y_train)
dt.fit(x_train_scaled, y_train)
```

```python
y_train_predict_knn = knn.predict(x_train_scaled)
y_train_predict_dt = dt.predict(x_train_scaled)
```
