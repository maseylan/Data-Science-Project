```python
import pandas as pd
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('/content/drive/MyDrive/kc_house_data.csv')
df.tail(10)
```

```python
total_null = df.isnull().sum()
print(total_null)
```

```python
df.info()
```

```python
# Mengubah nilai feature bathrooms dengan pembulatan ke atas
df['bathrooms'] = df['bathrooms'].apply(math.ceil)

# Mengubah tipe data kolom menjadi integer
df['bathrooms'] = df['bathrooms'].astype(int)
df.info()
```

```python
# Fitur numerik
numerical_features = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view',
                      'condition', 'grade', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
                      'lat', 'long', 'sqft_living15', 'sqft_lot15']

# Membuat scatter plot untuk setiap fitur numerik
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    plt.scatter(df.index, df[feature])
    plt.title(f'Scatter plot of {feature}')
    plt.xlabel('Index')
    plt.ylabel(feature)
    plt.show()
```
